```python
import csv
import statistics
from collections import Counter
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
```

# Given Data

```python
csv_file_path = './original.csv'
column_name = "Value"

data = []

with open(csv_file_path, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(float(row[column_name]))

data
```

# Figure 1.1

```python
plt.hist(data, bins=10, color='blue')
plt.xlabel(column_name, fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of ' + column_name, fontsize=14)
plt.grid(True)
plt.show()
```

# Figure 1.2

```python
np_data = np.array(data)
sm.qqplot(np_data, line='s')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.title('Q-Q Plot of ' + column_name)
plt.show()
```

### Stats

```python
mean = statistics.mean(data)

median = statistics.median(data)

data_counter = Counter(data)
mode_tuple = Counter(data).most_common(1)[0]
mode = mode_tuple[0]

std_deviation = statistics.stdev(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode_tuple[0]} appears {mode_tuple[1]} times")
print(f"Standard Deviation: {std_deviation}")
```

# Transformation

```python
from scipy.stats import boxcox

transformed_data, lambda_value = boxcox(data)
```

### Figure 2.1

```python
plt.hist(transformed_data, bins=10, color='blue')
plt.xlabel(column_name, fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of ' + column_name, fontsize=14)
plt.grid(True)
plt.show()
```

### Figure 2.2

```python
np_data = np.array(transformed_data)
sm.qqplot(np_data, line='s')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.title('Q-Q Plot of ' + column_name)
plt.show()
```

### Post-Transformation Stats

```python
mean = statistics.mean(transformed_data)

median = statistics.median(transformed_data)

data_counter = Counter(transformed_data)
mode_tuple = Counter(transformed_data).most_common(1)[0]
mode = mode_tuple[0]

std_deviation = statistics.stdev(transformed_data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode_tuple[0]} appears {mode_tuple[1]} times")
print(f"Standard Deviation: {std_deviation}")
```
