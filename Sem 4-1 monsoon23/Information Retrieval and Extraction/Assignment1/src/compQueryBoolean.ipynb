```python
# import numpy as np
# import pandas as pd
# from os import path
# from PIL import Image
# from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import os
import re
import math

import nltk
# from nltk.tokenize.toktok import ToktokTokenizer

# import matplotlib.pyplot as plt
%matplotlib inline
```

```python
corpus_path = "../corpus"

files = [os.path.join(corpus_path, filepath) for filepath in os.listdir(corpus_path)][:5]

data = []

# get data
for i, f in enumerate(files, 0):
	with open(f) as fo:
		curr = fo.read()
		data.append(curr)
```

```python
# remove spec chars and lower case it
data = [re.sub(r'[^\w\s\n]', ' ', doc).lower() for doc in data]
```

```python
# tokenise
data = [nltk.tokenize.ToktokTokenizer().tokenize(doc, return_str = True) for doc in data]
```

```python
# stem
stems_list = []
stemmer = nltk.PorterStemmer()
for idx, doc in enumerate(data):
	stems_list.append([stemmer.stem(word) for word in re.split(r"\s",doc)]) # type: ignore
stems_list
```

```python
vocab = []

for idx, doc in enumerate(stems_list):
	vocab.append({})
	for word in doc:
		if word in vocab[idx]:
			vocab[idx][word] +=1
		else:
			vocab[idx][word] = 1

vocab
```

# Boolean Model

```python
query = "thermal electron"
```

```python
result = []

for qw in re.split(r"\s", query):
	for idx, doc in enumerate(vocab):
		if (qw in doc) and (idx not in result):
			result.append(idx)

result
```
