```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import os
import re

import nltk
from nltk.tokenize.toktok import ToktokTokenizer

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
corpus_path = "../corpus"

files = [os.path.join(corpus_path, filepath) for filepath in os.listdir(corpus_path)][:5]

data = ""

# get data
for f in files:
	with open(f) as fo:
		curr = fo.read()
		data += "\n\n"
```

```python
# remove sepc chars and lower case it
data = re.sub(r'[^\w\s]', '', curr).lower()
```

```python
# tokenise
data = nltk.tokenize.ToktokTokenizer().tokenize(data, return_str = True)
```

```python
# stem
stemmer = nltk.PorterStemmer()

stems_list = [stemmer.stem(word) for word in data.split(" ")] # type: ignore
    
# Join our lists of words into a string again and assign to a variable.
data = ' '.join(stems_list)

print(data)
```

```python
freq_dist = nltk.FreqDist(stems_list)
freq_dist.plot(20, cumulative=False)
```

```python
wc = WordCloud(max_words=50).generate(data)
plt.imshow(wc)
```
